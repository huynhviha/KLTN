```python
from matplotlib import pyplot as plt
import numpy as np
from itertools import cycle

CLASS_NAMES = [ 'Atelectasis', 'Cardiomegaly', 'Effusion', 'Infiltration', 'Mass', 'Nodule', 'Pneumonia',
                'Pneumothorax', 'Consolidation', 'Edema', 'Emphysema', 'Fibrosis', 'Pleural_Thickening', 'Hernia']

def plot_roc_multi(roc_auc, fpr, tpr, classCount=5):
        
        lw = 2
        plt.figure(figsize=(10, 10))
        colors = cycle(['red', 'green', 'yellow', 'blue', 'pink', 'lime', 'tan', 'aqua', 'darkorchid', 'darkred', 'deeppink', 'darkorange', 'dimgray', 'plum'])
        for i, color in zip(range(classCount), colors):
            plt.plot(fpr[i], tpr[i], color=color, lw=lw,
                     label='{0} (area = {1:0.3f})'
                           ''.format(CLASS_NAMES[i], roc_auc[i]))

        plt.plot([0, 1], [0, 1], 'k--', lw=lw)
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.05])
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title('Receiver operating characteristic to multi-class')
        plt.legend(loc="lower right")
        plt.show()

roc_auc = np.load("result/densenet/roc_auc.npy")
fpr = np.load("result/densenet/fpr.npy")
tpr = np.load("result/densenet/tpr.npy")
plot_roc_multi(roc_auc, fpr, tpr, classCount=14)
```
